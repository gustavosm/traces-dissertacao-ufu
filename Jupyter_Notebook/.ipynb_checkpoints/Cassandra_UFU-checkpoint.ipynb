```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest, f_classif
```

```python
path        = 'Traces_Cassandra'
fields      = ['timestamp','W_95']
X_raw       = pd.read_csv(path+'/X_14_03.csv')
y_raw       = pd.read_csv(path+'/Y_14_03.csv', skipinitialspace=True, usecols=fields)
```

```python
a = plt.plot(y_raw['W_95'])
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_raw, y_raw['W_95'], test_size=0.3, random_state=11)
```

```python
regr = RandomForestRegressor(max_depth=10, random_state=17, n_estimators=120)
regr.fit(X_train, y_train)
```

```python
y_predicted = regr.predict(X_test)
#y_predicted = regr.predict(X_train)
```

```python
nmae_RF = (abs(y_predicted - y_test).mean())/y_test.mean()
#nmae_RF = (abs(y_predicted - y_train).mean())/y_train.mean()
nmae_RF
```

```python
plt.plot(y_test.values)
plt.plot(y_predicted)
plt.plot(abs(y_test.values-y_predicted))
plt.show()
```

```python
avg = y_train.mean()
error = y_test-avg
#nmae = (abs(error).mean())/avg
nmae_Naive = (abs(error).mean())/y_test.mean()
nmae_Naive
```

```python
selector = SelectKBest(f_classif, k=50)
regressor = RandomForestRegressor(max_depth=10, random_state=17, n_estimators=120)#, n_jobs=-1)#, random_state=10)

X_train_new = selector.fit_transform(X_train, y_train)
X_test_new = selector.transform(X_test)
#print(X_train_new.shape)
#print(X_train.shape)
#print(X_test_new.shape)
#print(X_test.shape)

idxs_selected = selector.get_support(indices=True)
print(idxs_selected)

regressor.fit(X_train_new, y_train)
pred = regressor.predict(X_test_new)
nmae_Selected = (abs(pred - y_test).mean())/y_test.mean()
nmae_Selected
```

```python
print("NMAE = ",nmae_RF)
print("NMAE Naive = ",nmae_Naive)
print("NMAE Selected = ",nmae_Selected)
```
